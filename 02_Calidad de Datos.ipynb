###### Variable beds

Vamos a utilizar un proxy que en este caso, va a ser la variable accomodates ya que podemos pensar como válido que el numero de personas de una vivienda es igual al numero de camas. Por ello, lo primero que haremos es filtrar por los registros que tienen datos en ambas variables (no filtramos los nulos) y ver si tienen algun patron válido que podamos usar luego para imputar.

```python
pd.crosstab(num.beds, num.accommodates)
```

Lo que observamos de esta tabla que la vemos en vertical es:
- Para una vivienda que permita el alojamiento de una persona, el máximo está en 1 cama.
- Para 2 personas, es 1 cama.
- Para 3 personas, son 2 camas
- Para 4 personas, son 2 camas.
- Para 5 personas, está entre 3 y 4 camas.

El patron que vamos a seguir es:
- Si el piso permite 1 o 2 personas, ponemos 1 cama.
- Si permite 3 o 4 personas, ponemos 2 camas.
- Si permite 5 o 6 personas, ponemos 3 camas.
- Si es más de 6 personas, ponemos 4 camas.

```python
def imputar_nulos(registro):
    #Lista de condiciones
    condiciones = [(registro.accommodates <= 2),
                  (registro.accommodates > 2) & (registro.accommodates <= 4),
                  (registro.accommodates > 4) & (registro.accommodates <=6),
                  (registro.accommodates > 6)]
    #lista de resultados
    resultados = [1,2,3,4]
    #Salida
    return (np.select(condiciones,resultados, default = -999))
```

```python
num.loc[num.beds.isna(), 'beds'] = num.loc[num.beds.isna()].apply(imputar_nulos, axis = 1).astype('int64')
```

### Atípicos

#### Por desviación típica

##### Identificación

###### Definir el número de desviaciones típicas

```python
num_desv_tip = 4
```

###### Función que devuelve los índices de los atípicos

```python
def atipicos_desv_tip(variable, num_desv_tip = 4):
    #sacamos los nulos por ahora
    variable = variable.dropna()
    #calculamos los límites
    media = np.mean(variable)
    sd = np.std(variable)
    umbral = sd * num_desv_tip
    lim_inf = media - umbral
    lim_sup = media + umbral
    #encontramos los índices de los que están fuera de los límites
    indices = [indice for indice,valor in variable.items() if valor < lim_inf or valor > lim_sup]
    return(indices)
```

###### Función que cuenta el número de atípicos

```python
def conteo_atipicos(df,variable, num_desv_tip = 3):
    atipicos = atipicos_desv_tip(df[variable], num_desv_tip)
    return(df.loc[atipicos,variable].value_counts())
```

###### Variables a aplicar

```python
var_atipicos_dt = num.columns.to_list()
```

###### Aplicar

```python
for variable in var_atipicos_dt:
    print('\n' + variable + ':\n')
    print(conteo_atipicos(num,variable,num_desv_tip))
```

#### Por winsorización automática para múltiples variables

##### Variables a winsorizar.

```python
var_winsorizar = ['price','minimum_nights','accommodates','bathrooms','bedrooms','beds']
```

##### Definir los percentiles mínimos y máximos.

```python
p_min = 0.01
p_max = 0.99
```

##### Aplicar

```python
# Winsorización automática para cada variable
for variable in var_winsorizar:
    print(f"Procesando la variable: {variable}")
    
    # Calcula los cuantiles para cada variable
    lower_quantile = num[variable].quantile(p_min)
    upper_quantile = num[variable].quantile(p_max)
    
    # Si la columna es de tipo Int64, redondea los cuantiles a enteros
    if pd.api.types.is_integer_dtype(num[variable]):
        lower_quantile = int(round(lower_quantile))
        upper_quantile = int(round(upper_quantile))
        print(f"Cuantiles redondeados para la variable de tipo Int64 '{variable}': {lower_quantile}, {upper_quantile}")
    else:
        print(f"Cuantiles para la variable de tipo Float '{variable}': {lower_quantile}, {upper_quantile}")
    
    # Aplica la winsorización utilizando los cuantiles como límites
    num[variable].clip(lower=lower_quantile, upper=upper_quantile, inplace=True)

    # Imprimir los resultados después de la winsorización
    print(f"Resultados después de la winsorización para la variable '{variable}':")
    print(num[variable].describe())
```
